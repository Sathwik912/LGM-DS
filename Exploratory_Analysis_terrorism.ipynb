```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
     
```

```python
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
data.head()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
data['Casualities'] = data['kill'] + data['Wound']
data.info()
```

```python
data.describe()
```

# Data Visualization

```python
years = data['Year'].unique()
attack_counts = data['Year'].value_counts(dropna=False).sort_index()

plt.figure(figsize=(14, 8))
sns.barplot(x=years, y=attack_counts, palette = 'RdPu')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Number of Terrorist Attacks Each Year')
plt.show()
```

```python


plt.figure(figsize=(50, 30))
crosstab_data = pd.crosstab(data.Year, data.Region)
crosstab_data.plot(kind='area', stacked=False, cmap='magma', alpha=0.9)
plt.title('Terrorist Activities By Region In Each Year', fontsize=15)
plt.ylabel('Number of Attacks', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), fontsize=10)
sns.set_style('darkgrid')
plt.tight_layout()
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(x=data['Country'].value_counts()[:15].index,
            y=data['Country'].value_counts()[:15].values,
            palette='Blues_d')
plt.title('Top 15 Countries Affected by Terrorism', fontsize=14)
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Number of Attacks', fontsize=12)
plt.xticks(rotation= 45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("Number of People Killed in Terrorist Attacks",fontsize=25)
plt.ylabel("Number of people killed ",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
data['Attacktype'].value_counts().plot(kind='bar', figsize=(20,10), color=sns.color_palette('Set2', n_colors=9))
plt.xticks(rotation=50, fontsize=12)
plt.xlabel("Attack Type", fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.title("Types of Attacks", fontsize=20)
plt.show()
```

```python
df = data[['Group', 'Country', 'kill']]
df = df.groupby(['Group', 'Country'], axis=0).sum().sort_values('kill', ascending=False).drop('Unknown').reset_index().head(10)

plt.figure(figsize=(20, 10))
sns.barplot(x='kill', y='Group', hue='Country', data=df, palette='Paired')
plt.title('Top 10 Terrorist Groups causing highest number of Deaths', fontsize=20)
plt.xlabel('Number of Deaths', fontsize=15)
plt.ylabel('Terrorist Group', fontsize=15)
plt.legend(loc='upper right', fontsize=15)
plt.show()
```

Report on Global Terrorism Trends (1970-2017)

Key Findings:



*   Iraq had the highest number of terrorist attacks, followed by Pakistan and Afghanistan.
*   Baghdad was the city with the highest number of terrorist attacks


*   The Middle East & North Africa region was the most affected by terrorist attacks.

*   The year 2014 had the most number of terrorist attacks, with the month of May being the most affected.

*   The Taliban was the terrorist group responsible for the most attacks
*   Bombing/Explosion was the most common type of attack.




*   A total of 181,691 terrorist attacks were recorded from 1970 to 2017, causing about 935,737 casualties, which consisted of 411,868 deaths.

Conclusion:
Terrorism remains a major global threat with serious consequences. Our analysis shows that the Middle East and North Africa region is the most affected, with Iraq being the most affected country, followed by Pakistan and Afghanistan. Bombing and explosions are the most common types of attacks, and the Taliban is the group responsible for the most attacks. Understanding these trends and patterns is crucial for policymakers and security agencies to develop effective strategies to combat terrorism.
